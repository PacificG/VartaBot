```python
districts = df['District'].value_counts().index
```

```python
f = open("district.txt", "w")
ls = [district for district in districts]
for district in districts:
    for s in district.split(" "):
        if s not in ls:
            ls.append(s)
f.writelines([l + ", " for l in ls])
f.close()
```
